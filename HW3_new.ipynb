```python
import cv2
import numpy as np
import time
import matplotlib.pyplot as plt
start=time.time()
org_img=cv2.imread(r"C:input\cropHSV_yellow.jpg",-1) 
hsvimg =cv2.cvtColor(org_img, cv2.COLOR_RGB2HSV)
plt.title("HW3")
plt.imshow(org_img)
plt.show()
print(time.time()-start)


med = cv2.medianBlur(hsvimg, 3)
'''

lowerb = np.array([0,0,0])
upperb = np.array([180, 255, 255])

'''
lowerb = np.array([110,100 , 50])
upperb = np.array([180, 255, 255])
maskmed = cv2.inRange(org_img, lowerb, upperb)
plt.imshow(maskmed,cmap='gray')
plt.title("HW3")
plt.show()
```

```python

kernel = np.ones((7, 7), np.uint8)
dilated_img = cv2.dilate(maskmed, kernel, iterations=1)
plt.plot()
plt.imshow(dilated_img, cmap='gray')
plt.title("Dilated")
plt.show()
eroded_img = cv2.erode(dilated_img, kernel, iterations=1)
plt.plot()
plt.imshow(eroded_img, cmap='gray')
plt.title("Eroded")
plt.show()
```
